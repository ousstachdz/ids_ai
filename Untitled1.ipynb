```python
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution1D, MaxPooling1D
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder
```

```python
import urllib.request


url = "http://archive.ics.uci.edu/ml/machine-learning-databases/kddcup99-mld/kddcup.data.gz"
filename = "kddcup.data.gz"

with urllib.request.urlopen(url) as response, open(filename, 'wb') as out_file:
    while True:
        data = response.read(1024)
        if not data:
            break
        out_file.write(data)
```

```python
chunk_size = 1_000_000
for chunk in pd.read_csv('kddcup.data', chunksize=chunk_size):
    X = chunk.iloc[:, :-1].values.reshape(len(chunk), -1)
    y = chunk.iloc[:, -1].values

num_classes = len(set(y))
num_classes
```

```python
np.unique(y)

label_dict = {'back.': 0, 'buffer_overflow.': 1, 'ipsweep.': 2, 'land.': 3, 'neptune.': 4, 'normal.': 5, 'phf.': 6, 'pod.': 7, 'portsweep.': 8, 'rootkit.': 9, 'satan.': 10, 'smurf.': 11, 'teardrop.': 12}
y_encoded = [label_dict[label] for label in y]
```

```python
df_y = keras.utils.to_categorical(y_encoded,num_classes=num_classes)
```

```python
df_x = np.array(X)
df_y = np.array(df_y)
```

```python
df_y
```

```python
df_x.shape
```

```python
label_encoder1 = LabelEncoder()
label_encoder2 = LabelEncoder()
label_encoder3 = LabelEncoder()

# fit each encoder to the corresponding column of string data
label_encoder1.fit(df_x[:,1])
label_encoder2.fit(df_x[:,2])
label_encoder3.fit(df_x[:,3])

# transform each column of string data into encoded numeric values
df_x[:,1] = label_encoder1.transform(df_x[:,1])
df_x[:,2] = label_encoder2.transform(df_x[:,2])
df_x[:,3] = label_encoder3.transform(df_x[:,3])
```

```python
# print(label_encoder1.classes_)
# print(label_encoder2.classes_)
print(label_encoder3.classes_)
```

```python
df_y[0]
```

```python
df_x[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
x_train[0]
```

```python
x_test[0]
```

```python
x_train.shape
```

```python
y_train[0]
```

```python
y_test[0]
```

```python
cnn = Sequential()
cnn.add(Convolution1D(64, 3, activation="relu", input_shape=(41, 1)))
cnn.add(MaxPooling1D(pool_size=2))
cnn.add(Flatten())
cnn.add(Dense(128, activation="relu"))
cnn.add(Dropout(0.5))
cnn.add(Dense(13, activation="softmax"))
cnn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
cnn.fit(x_train.astype('float32'),y_train.astype('float32'),validation_data=(x_test.astype('float32'),y_test.astype('float32')))
```

```python
cnn.evaluate(x_test.astype('float32'),y_test.astype('float32'))
```

```python
cnn.save('./model/cnn_ids.hdf5')
```
